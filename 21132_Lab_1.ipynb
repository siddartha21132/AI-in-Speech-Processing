```python

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

A1

```python
new_audio_data, new_sample_rate = librosa.load('Siddartha.wav')
plt.figure(figsize=(8, 4))
librosa.display.waveshow(new_audio_data)
plt.title('New Speech Signal Waveform')
plt.show()
```

A2

```python
print("Length of the signal:", len(new_audio_data), "samples")
print("Sampling rate:", new_sample_rate, "Hz")
print("Magnitude range:", np.min(new_audio_data), "to", np.max(new_audio_data))
```

A3

```python
import IPython.display as ipd

start_time = int(1.275 * new_sample_rate)
end_time = int(3.755 * new_sample_rate)
custom_segment = new_audio_data[start_time:end_time]

plt.figure(figsize=(8, 5))
librosa.display.waveshow(custom_segment)
plt.title('Custom Segmented Audio')
plt.show()
print("Custom Segmented Audio:")
ipd.Audio(custom_segment, rate=new_sample_rate)
```
